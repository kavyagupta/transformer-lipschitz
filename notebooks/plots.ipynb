```python
import os
import glob
import seaborn as sns 

%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
```

```python
logs = glob.glob('../logs/*mnist_seed*layers-6*warmup*L2.csv')
```

```python
relax_warmup_seed_map = {}
for log in logs:
    df = pd.read_csv(log, header=None)
    name = log.split('/')[-1][:-4]
    _, seed, _, relax, warmup, _ = name.split('-')
    relax = float(relax.split('_')[0])
    seed = int(seed.split('_')[0])
    warmup = int(warmup.split('_')[0])
    post_warmup = df[warmup:]
    best_epoch = post_warmup[:][1].idxmax()
    relax_warmup_seed_map[(seed, relax, warmup)] = post_warmup.iloc[best_epoch-warmup]
```
